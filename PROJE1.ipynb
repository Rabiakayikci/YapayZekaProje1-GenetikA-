```python
#genetik_operators.py

# Kodlar metin iÃ§ine yazÄ±lÄ±r ve py uzantÄ±lÄ± bir dosya olarak kaydedilir.

kod = '''

import numpy as np
import random


def birey_olustur ():
   x1 = np.random.uniform (25,80) #25 ten baÅŸlattÄ±m Ã§Ã¼nkÃ¼ x1 >25 koÅŸulu var eksi puan yememesi iÃ§in

   x2 = np.random.uniform (10,80)

   birey = np.array([x1,x2])
   return birey


def populasyon_olustur (populasyon_sayisi):
  pop= []

  for i in range (populasyon_sayisi):
    yeni = birey_olustur()
    pop.append(yeni)

  return pop


def tahmin_y (birey): #aÄŸÄ±rlÄ±klarÄ±mÄ± fonksiyona koyup y deÄŸerini hesaplÄ±yorum

    #benim fonksiyonum 2 deÄŸiÅŸkenli olduÄŸu iÃ§in 2 birey olacak, dizi ÅŸeklinde (x1,x2)
    x1= birey[0]
    x2= birey[1]

    y = 3*x1+ 2*x2+ x1*x2- 0.5 *(x2**2)
    return y


def kontrol (birey): #aykÄ±rÄ± deÄŸerleri hesaplayÄ±p eksiltme yapacaÄŸÄ±z
    x1= birey[0]
    x2= birey[1]

    hata_sayisi = 0
    if x1 < 25:
      hata_sayisi +=1

    if x1 + x2 > 100:
      hata_sayisi += 1

    return hata_sayisi

def uygunluk_hesapla (birey):

    as_puan = tahmin_y(birey)
    hata_tekrari = kontrol(birey)

    ceza_puani = hata_tekrari * 3000

    gercek_uyg = as_puan - ceza_puani
    return gercek_uyg


def rulet_secim (populasyon,uygunluklar, adet=2):

  uygunluklar = np.array(uygunluklar)

  min_puan = np.min(uygunluklar)

  if min_puan < 0:
      poz_uygunluklar = uygunluklar - min_puan +1    #rulette negatif deÄŸerler sÄ±kÄ±ntÄ± Ã§Ä±karabileceÄŸi iÃ§in dizideki en kÃ¼Ã§Ã¼k deÄŸeri 1 yapacak ÅŸekilde pozitife dÃ¶ndÃ¼rme yapÄ±yoruz

  else:
     poz_uygunluklar= uygunluklar

  toplam_puan = np.sum(poz_uygunluklar)

  if toplam_puan == 0:
      rastegele_indeksler = np.random.choice(len(populasyon), size = adet) #populasyonun iÃ§inden adet kadar random bireyin indeksini seÃ§er
      return populasyon[rastegele_indeksler]

  olasiliklar = poz_uygunluklar / toplam_puan
  secilen_indeksler = np.random.choice(len(populasyon), size = adet, p = olasiliklar) # hesaplanan olasÄ±klara gÃ¶re populasyondan seÃ§im yapar
  return populasyon[secilen_indeksler]


def rank_secim (populasyon, uygunluklar, adet =2):
  N= len(populasyon)
  sirali_indeks = np.argsort(uygunluklar)  #Puanlara gÃ¶re kÃ¼Ã§Ã¼kten bÃ¼yÃ¼ÄŸe sÄ±ralama yapar(ADRESLERÄ° SIRALAR)
  payda = N * (N+1) / 2
  secim_ihtimal = np.array([(i+1)/ payda for i in range(N)]) #ranklarÄ±n hepsi toplandÄ± ve paydaya eÅŸitlendi,pay kÄ±smÄ± ise hangi rankta olduÄŸunu ifade eder o yÃ¼zden dÃ¶ngÃ¼ye sokarak olasÄ±lÄ±klarÄ±nÄ± hespladÄ±k

  ihtimaller = np.zeros(N)
  for i, orjinal_indeks in enumerate(sirali_indeks): #enumerate bize 2 deÄŸer verir
                                                            #1.dÃ¶ngÃ¼ kaÃ§Ä±ncÄ± adÄ±mda
                                                            #2.orjinal dizide neredeyiz

    ihtimaller[orjinal_indeks] = secim_ihtimal[i]  # kÃ¼Ã§Ã¼kten bÃ¼yÃ¼k deÄŸere gÃ¶re sÄ±ralanmÄ±ÅŸ adresleri sÄ±rayla gezer ve onlarÄ±n olasÄ±lÄ±klarÄ±nÄ± dÃ¼zenli bir ÅŸekilde ihtimaller dizisine yazar
  ihtimaller = ihtimaller / np.sum(ihtimaller)#olasÄ±lÄ±k deÄŸerlerinin toplamÄ± 1 mi kontrol eder

  secilen_indeks = np.random.choice(len(populasyon), size = adet, p = ihtimaller)
  return populasyon[secilen_indeks]


def tek_noktali_cap(p1,p2):
  nokta = random.randint(1, len(p1)-1) #diziyi (p1 i)1'den baÅŸlayarak uzunluÄŸundan 1 eksik kadar bÃ¶ler (bizim dizi 2 eleman olduÄŸu iÃ§in ikiye bÃ¶lmÃ¼ÅŸ oluyor) randint => bÃ¶lme komutu
                                       #nokta deÄŸiÅŸkeni bir indis sayÄ± tutar

  c1 = np.concatenate((p1[:nokta], p2[nokta:])) #p1[:nokta] => 0. indisten baÅŸla nokta indisine kadar git nokta indisini alma p2[nokta:] => nokta indisinden baÅŸla sonuna kadar al
                                                #concatenate => birleÅŸtir bunlarÄ±
  c2 = np.concatenate((p2[:nokta], p1[nokta:]))

  return c1, c2

#benim uygulamam 2 veri kÃ¼mesinden (x1,x2) oluÅŸuyor bu yÃ¼zden iki noktalÄ± Ã§aprazlama yapamam bir bireyi baÅŸka bir ÅŸekilde bÃ¶lemem


def mutasyon(birey, ihtimal, buyukluk):
  yeni_birey = birey.copy()

  for i in range(len(yeni_birey)):
      if np.random.rand() < ihtimal:
        degisim = buyukluk * (np.random.rand()-0.5)
        yeni_birey[i] += degisim

  yeni_birey[0] = np.clip(yeni_birey[0], 25, 80) #aykÄ±rÄ± deÄŸer olmamasÄ± yani x1 > 25 x1<80 olma koÅŸulunu aÅŸmamasÄ± iÃ§in dÃ¼zenleme yapÄ±lÄ±r eÄŸer 25'ten kÃ¼Ã§Ã¼kse 25 olur
  yeni_birey[1] = np.clip(yeni_birey[1], 10, 80)
  return yeni_birey

'''
with open("genetik_operators.py", "w") as dosya:
    dosya.write(kod)

print("âœ… genetik_operators.py baÅŸarÄ±yla oluÅŸturuldu.")


```

```python
kod = """
import numpy as np
import random
import matplotlib.pyplot as plt
from genetik_operators import (
    birey_olustur,
    populasyon_olustur,
    tahmin_y,
    kontrol,
    uygunluk_hesapla,
    rulet_secim,
    rank_secim,
    tek_noktali_cap,
    mutasyon
)

def evrimsel_algoritma(populasyon, nesil_sayisi,
                        caprazlama_turu, secim_turu, mutasyon_ihtimali, mutasyon_buyuklugu):

    en_iyiler = []

    for nesil in range(nesil_sayisi):

        print(f"\\nğŸ” NESÄ°L {nesil+1}")

        uygunluklar = np.array([uygunluk_hesapla(b) for b in populasyon])

        en_iyi_indeks = np.argmax(uygunluklar)
        elit_birey = populasyon[en_iyi_indeks].copy()
        elit_uygunluk = uygunluklar[en_iyi_indeks]

        en_iyiler.append(elit_uygunluk)

        print("Uygunluklar:", np.round(uygunluklar, 3))

        havuz = []
        for _ in range(len(populasyon)// 2):

            if secim_turu == "rulet":
                secilenler = rulet_secim(populasyon, uygunluklar, adet=2)
            elif secim_turu == "rank":
                secilenler = rank_secim(populasyon, uygunluklar, adet = 2)
            else:
                 raise ValueError("Hata: SeÃ§im tÃ¼rÃ¼ 'rulet' veya 'rank' olmalÄ±.")

            havuz.append(secilenler)

        print("\\nğŸ‘¥ EÅŸleme Havuzu:")

        for i, (p1, p2) in enumerate(havuz):
           print(f"  Ebeveyn {2*i+1}: {np.round(p1,1)}")
           print(f"  Ebeveyn {2*i+2}: {np.round(p2,1)}")


        yeni_bireyler = [elit_birey]

        while len(yeni_bireyler) < len(populasyon):
            p1, p2 = np.array(random.choice(havuz))
            if caprazlama_turu == "tek":
               c1, c2 = tek_noktali_cap(p1, p2)

            else:
               raise ValueError("Ã‡aprazlama tÃ¼rÃ¼ 'tek' olmalÄ±.")

            c1 = mutasyon(c1, mutasyon_ihtimali, mutasyon_buyuklugu)
            c2 = mutasyon(c2, mutasyon_ihtimali, mutasyon_buyuklugu)
            yeni_bireyler.extend([c1, c2])

        populasyon = np.array(yeni_bireyler[:len(populasyon)])

        print("\\nğŸ† En iyi birey:", np.round(elit_birey, 1))
        print("Tahmini y:", tahmin_y(elit_birey).round(2))
        print("Uygunluk:", round(elit_uygunluk, 4))

        # Grafik Ã§izgisini gÃ¶rmek iÃ§in hedefi 100.000 yaptÄ±k
        if elit_uygunluk >= 1e5:
            print("âœ… Hedefe ulaÅŸÄ±ldÄ±. Erken durdurma uygulandÄ±.")
            break

    # Grafik Ã‡izimi (Buradaki girinti hatasÄ± dÃ¼zeltildi)
    plt.figure(figsize=(10,6))
    plt.plot(en_iyiler, marker='o', linestyle='-', color='b')
    plt.title("En Ä°yi Uygunluk DeÄŸeri (Fitness) DeÄŸiÅŸimi")
    plt.xlabel("Nesil")
    plt.ylabel("Uygunluk (Fitness)")
    plt.grid(True)
    plt.show()
"""

with open("evrim_motoru.py", "w") as dosya:
    dosya.write(kod)

print("âœ… evrim_motoru.py baÅŸarÄ±yla oluÅŸturuldu.")
```
